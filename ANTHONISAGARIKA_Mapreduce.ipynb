Creating two seperate python functions mapper.py and reducer.py to count number of words


The below python code is for mapper:
1)It does encoding the given file
2)Removing all the punctuations 
3)Removing white spaces 
4)And generating key value pairs


```python
import re
import string

# Define the filename
filename = 'file1.txt'  # Replace with the actual filename or path if needed

# Open the file for reading with 'utf-8' encoding
with open(filename, 'r', encoding='utf-8') as file:
    for line in file:
        line = line.lower()
        
        # Use regex to remove punctuation and digits
        line = re.sub(r"[%s%s]" % (re.escape(string.punctuation), re.escape(string.digits)), " ", line)
        
        # Using regex to split the line into words
        words = re.findall(r'\b\w+\b', line)
        
        # Emit key-value pairs for each word
        for word in words:
            print(f"{word}\t{filename}:1")
            
```

```python
#Creating a reducer program to find the number of word occurances
```

```python
# Initialize a dictionary to store word occurrences
word_count_dict = {}
for word in words:
            # If the word is not in the dictionary, add it with a count of 1
            if word not in word_count_dict:
                word_count_dict[word] = 1
            else:
                # If the word is already in the dictionary, increment its count
                word_count_dict[word] += 1

# Print the word count dictionary
for word, count in word_count_dict.items():
    print(f"{word}: {count}")


```

From the second text file (file2.txt), write Python code and use MapReduct to count how many times non-English words (names, places, spells etc.) were used. List those words and how many times each was repeated.

```python
#using the same mapper function for the second file too
```

```python
import re
from collections import defaultdict
from functools import reduce

# Load the list of English words from the file extracted from (http://www.gwicks.net/dictionaries.htm) 
english_words_file = 'usa.txt'
with open(english_words_file, 'r', encoding='utf-8') as file:
    english_words = set(word.strip().lower() for word in file.readlines())

def is_english(word):
    return word.lower() in english_words

# Mapper function to remove unwanted characters and count non-English words
def map_non_english_words(line):
    # Use regular expressions to remove unwanted characters adjacent to words
    words = re.findall(r'\b\w+\b', line.lower())
    non_english_words = {}
    for word in words:
        if not is_english(word):
            non_english_words[word] = non_english_words.get(word, 0) + 1
    return non_english_words
```

```python
#Reduce function is used to count the occurances of non english words
```

```python
def reduce_non_english_words(count1, count2):
    for word, count in count2.items():
        count1[word] = count1.get(word, 0) + count
    return count1

with open('file2.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

mapped_non_english_words = list(map(map_non_english_words, lines))

final_non_english_words = defaultdict(int)
for word_count in mapped_non_english_words:
    final_non_english_words = reduce(reduce_non_english_words, [final_non_english_words, word_count])

for word, count in final_non_english_words.items():
    print(f'{word}: {count}')
```
